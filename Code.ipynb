```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import logging
import csv
```

```python
article_url = "https://www.nytimes.com/"
```

```python
article_url
```

```python
urlclient = urlopen(article_url)
urlclient
```

```python
article_page = urlclient.read()
article_page
```

```python
article_html = bs(article_page, 'html.parser')
article_html
```

```python
bigbox = article_html.findAll("li" , {"class":"css-1d8trz3"})
```

```python
len(bigbox)
```

```python
bigbox = bigbox[0:11]
bigbox
```

```python
newslink = bigbox[0].a['href']
```

```python
news_req = requests.get(newslink)
```

```python

urls = []
for i in bigbox:
    urls.append(i.a['href'])
```

```python
urls
```

```python

# Variable to store scraped data
scraped_data = []

for url in urls:
    for page_number in range(1, 4):
        
        page_url = f'{url}?page={page_number}'

        response = requests.get(page_url)
        
        if response.status_code == 200:
            
            soup = bs(response.text, 'html.parser')
            
            
            h3_tags = soup.find_all('h3', {'class': 'css-1kv6qi e15t083i0'})
            
            for h3_tag in h3_tags:
                scraped_data.append(f"{url.split('/')[-1]}, '{h3_tag.text.strip()}'")
print('Scraped data:', scraped_data)
```

```python
with open('output.csv', 'w', newline='') as csvfile:
    fieldnames = ['section', 'News_title']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()

    for item in scraped_data:
       
        section, News_title = item.split(', ', 1)

        
        News_title = News_title.strip("'")

        
        writer.writerow({'section': section, 'News_title': News_title})

print("CSV conversion complete. Check 'output.csv'.")
```
